<h1 align="center">Annual income prediction</h1>   
  
This data was extracted from the census bureau database found at  
 http://www.census.gov/ftp/pub/DES/www/welcome.html  
 
  
 Prediction task is to determine whether a person makes over 50K a year.  


```python
# set working directory
import os
os.getcwd()
os.chdir('/data')

# import packages
import sys
import numpy as np
import pandas as pd
from xgboost.sklearn import XGBClassifier
import xgboost as xgb
import matplotlib.pyplot as plt
%matplotlib inline

# set seed
np.random.seed(0)

# read data
df_train = pd.read_csv('adult.data.txt', header=0)
df_test = pd.read_csv('adult.test.txt',header=0)

# add column names
df_train.columns = ['age', 'workclass','fnlwgt','education','education-num','marital-status','occupation','relationship',
                   'race','sex','capital-gain','capital-loss','hours-per-week','native-country','50k']
df_test.columns = ['age', 'workclass','fnlwgt','education','education-num','marital-status','occupation','relationship',
                   'race','sex','capital-gain','capital-loss','hours-per-week','native-country','50k']
```

```python
# number of training samples
num_train = df_train.shape[0]

# make training/testing levels consistent(ie: <50k to <50k.)
df_train['50k'] = df_train['50k'].map(lambda x: str(x)+'.')

# merge training/testing data
df_all = pd.concat((df_train, df_test), axis=0, ignore_index=True)
```

<h2 align='center'> Constructing Model </h2>

```python
# expand categorial variables
categoricalFeatures = [ 'workclass','education','marital-status','occupation','relationship',
                   'race','sex','native-country','50k']

# transform to dummy variable
df_all_dummy = pd.get_dummies(df_all[categoricalFeatures])

# combine with df_all
df_all = pd.concat((df_all, df_all_dummy), axis=1)
df_all = df_all.drop(categoricalFeatures, axis=1)
df_all.info()
```

```python
# extract target
target = df_all['50k_ >50K.'].values
df_all = df_all.drop(['50k_ >50K.','50k_ <=50K.'], axis=1)
df_all.info()
```

```python
# select features
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_classif
Selector_f = SelectPercentile(f_classif, percentile=25)
Selector_f.fit(df_all,target)
for n,s in zip(df_all.columns,Selector_f.scores_):
    print('F-score: %8.2f %s ' % (s,n))
```

```python
Selector_f.transform(df_all).shape
```

```python
# training/testing matrix
dtrain = xgb.DMatrix(Selector_f.transform(df_all)[num_train:,], label = target[num_train:])
dtest = xgb.DMatrix(Selector_f.transform(df_all)[num_train:,], label = target[num_train:])

# specify parameters
param = {'max_depth':6, 'eta':0.4, 'silent':0, 'objective':'binary:logistic', 'n_estimators':20, 'eval_metric':'error'}

num_round = 50
```

```python
# test for parameters using cross validation
cv_xgb = xgb.cv(param, dtrain, num_round, nfold=5,seed = 0)
print(cv_xgb)
```

```python
# train model
model_xgb = xgb.train(param, dtrain, 20)
```

```python
model_xgb.get_fscore
```

```python
model_xgb.dump_model('model_xgb.txt')

with open('model_xgb.txt') as f: 
    print(f.read()) 
```

<h2 align = 'center'> Prediction </h2>

```python
# predict for dtest
y_pred = model_xgb.predict(dtest)

# error rate for perdiction
np.mean((target[num_train:]-(y_pred>0.5))**2)
```
